```python
from litellm import completion
from typing import Any, Optional

def single_shot_llm_call(
    model: str,
    system_prompt: str,
    message: str,
    response_format: Optional[dict[str, str | dict[str, Any]]] = None,
    max_completion_tokens: int | None = None,
) -> str:
    response = completion(
        model=model,
        messages=[{"role": "system", "content": system_prompt}, {"role": "user", "content": message}],
        temperature=0.0,
        response_format=response_format,
        max_tokens=max_completion_tokens,
        timeout=1000,
    )
    
    return response.choices[0].message["content"]  # type: ignore
```

```python
answer = ["token"]*10000
answer = " ".join(answer)
print(single_shot_llm_call("together_ai/Qwen/Qwen2.5-72B-Instruct-Turbo", "You are a helpful assistant.",f"{answer}\nWhat is the capital of France?"))
```
